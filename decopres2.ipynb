#What are decorators?

* Defined in PEP 318 (Python 2.4)
* Callables that:
    * Take a callable
    * Do something
    * Return a callable

#Why?

* Add or remove logic from functions
* Registration
* Pre- or Post-processing
* Callbacks
* Many Others!

##Syntax

```python
def awesome(f):
    return f

@awesome
def my_func():
    pass

my_func = awesome(my_func)
```

##A Basic Decorator

```python
from time import time
from functools import wraps

def timeit(f):
    @wraps(f)
    def timer(*args, **kwargs):
        start = time()
        res = f(*args, **kwargs)
        print(time() - start)
        return res
    return timer

@timeit
def hello():
    print("Hello!")
    
hello()
```

##Pass through decorators

```python
def make_awesome(f):
    f.awesome = True
    return f

@make_awesome
def thing():
    return "Thing()"

assert hasattr(thing, "awesome")
```

##Flexible Closures

```python
def brittle(f):
    def wrapper(arg1, arg2, kwarg1=None, kwarg2=None):
        return f(arg1, arg2, kwarg1, kwarg2)
    return wrapper
```

```python
def flexible(f):
    def wrapper(*args, **kwargs):
        return f(*args, **kwargs)
    return wrapper
```

###Important because...

```python
@flexible
def dinner(food, beer): # actually becomes wrapper
    print("Eating", food, "and drinking", beer)

dinner("Bratzel", "Azazel")
print(dinner.__name__)
```

##Preserving Metadata

###Move it over manually

```python
def loggit(f):
    def logger(*args, **kw):
        print("Calling {}.".format(f.__name__))
        return f(*args, **kw)
    logger.__name__ = f.__name__
    logger.__doc__ = f.__doc__
    logger.__dict__.update(f.__dict__)
    return logger
```

###Python can do this for us

```python
from functools import wraps

def loggit(f):
    @wraps(f)
    def logger(*args, **kwargs):
        print("Calling {}.".format(f.__name__))
        return f(*args, **kwargs)
    return logger

@loggit
def random():
    """Chosen by fair dice roll."""
    return 4

print(random.__name__)
print(random.__doc__)
```

###What is @wraps?

* `wraps` uses `update_wrapper`
* `update_wrapper` moves metadata from one object to another
    * `__name__`, `__doc__`, `__module__`, `__dict__`
    * `__qualname__` and `__annotations__` in Py3
    * `__wrapped__` in >= 3.2

##Optional kwargs

###Probably the most common pattern

```python
def before(phrase="Setting up"):
    def actual_deco(f):
        @wraps(f)
        def wrapper(*args, **kwargs):
            print(phrase)
            return f(*args, **kwargs)
        return wrapper
    return actual_deco
```

```python
@before(phrase="much setup!")
def does_nothing():
    print("I did nothing.")
    
@before() # parens must be present
def add_two(x):
    return x + 2
```

###Why not a partial?

```python
from functools import partial

def before(f=None, phrase="Setting up"):
    if f is None:
        return partial(before, phrase=phrase)
    
    @wraps(f)
    def wrapper(*args, **kwargs):
        print(phrase)
        return f(*args, **kwargs)
    
    return wrapper
```

```python
@before(phrase="much setup!")
def does_nothing():
    print("I did nothing.")
    
@before # parens not required
def add_two(x):
    return x + 2
```

###Decorators all the way Down

```python
def optional_kwargs(deco):
    @wraps(deco)
    def wrapper(f=None, **kwargs):
        if f is None:
            return partial(wrapper, **kwargs)
        return deco(f, **kwargs)
    return wrapper
```

```python
@optional_kwargs
def before(f, phrase="Setting up"):
    @wraps(f)
    def wrapper(*args, **kwargs):
        print(phrase)
        return f(*args, **kwargs)    
    return wrapper

@before
def add_two(x):
    return x + 2

add_two(4)
```

##Decorating Classes

* Introduced in Python 2.6
* PEP 3129

```python
from functools import total_ordering

@total_ordering
class Frob:
    
    def __init__(self, bar):
        self.bar = bar
        self.baz = bar * 2
    
    def __eq__(self, other):
        (self.bar + self.baz) == (other.bar + other.baz)
    
    def __gt__(self, other):
        (self.bar + self.baz) > (other.bar + other.baz)
        
Frob(4) < Frob(2)
```

##Using Objects To Decorate

* One instance per decoration
* One instance for all decoration

```python
from functools import update_wrapper

class wrapperobj:
    def __init__(self, f):
        update_wrapper(self, f)
        self.f = f
    
    def __call__(self, *args, **kwargs):
        return self.f(*args, **kwargs)

@wrapperobj
def random():
    return 4
print(random())
```

```python
class singlewrapper:
    
    def __call__(self, f):
        @wraps(f)
        def wrapper(*args, **kwargs):
            return f(*args, **kwargs)
        return wrapper

wrapper = singlewrapper()

@wrapper
def random():
    return 4
print(random())
```

##Decorating Methods

* Easy with functional closures
* Harder with objects

```python
class Foo:
    @loggit
    def frob(self, n=2):
        print(" ".join(["Frob"] * n))

Foo().frob(n=2)
```

```python
class Foo:
    @wrapperobj
    def frob(self, n=1):
        print(" ".join(["Frob"] * n))

Foo().frob
```

###Non-Data Descriptors In One Slide

* Instances that live on classes
* `__get__` serves as the closure
* Partially apply the method
* Python handles the rest

```python
class methodwrapper:
    def __init__(self, method):
        self.method = method
    
    def __get__(self, instance, cls):
        if instance is None:
            return self.method
        return partial(self.method, instance)
```

```python
class Foo:
    @methodwrapper
    def frob(self, n=1):
        print(" ".join(["Frob"] * n))

Foo().frob(n=2)
```
